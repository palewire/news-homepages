```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("../extracts/csv/screenshot-files.csv", parse_dates=["mtime"])
```

```python
df.info()
```

```python
df['date'] = df.mtime.dt.date
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
sites_by_date = df[['date', 'handle']].drop_duplicates().groupby("date").size().rename("sites").reset_index()
```

```python
sites_by_date
```

```python
alt.Chart(sites_by_date).mark_bar().encode(
    x=alt.X("date:T"),
    y=alt.Y("sites:Q"),
)
```

```python
sites_by_date.to_csv("sites_by_date.csv", index=False)
```
